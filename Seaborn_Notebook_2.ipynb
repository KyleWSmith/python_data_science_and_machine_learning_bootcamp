# Categorical Plots

---



```
import seaborn as sns
%matplotlib inline
#Set figure output size:
sns.set(rc={'figure.figsize':(11.7,8.27)})
tips = sns.load_dataset('tips')
tips.head()
```

```
tips.describe()
```

```
import numpy as np
```

```
sns.barplot(x='sex' , y = 'total_bill' , data=tips, estimator=np.std)
```

```
#Count plot
sns.countplot(x='sex', data=tips)
```

```
sns.boxplot(x= 'day',y='total_bill',data=tips, hue='smoker')
```

```
#Violin Plot

sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',split=True, )
```

```
#Strip Plot
#x categorical, y numerical for these plots
#Jitter makes it easier to read

sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex',)
```

```
#Swarm plot - points are adjusted so they don't overlap.
#Drawback is that they don't scale well for large numbers of points.

sns.swarmplot(x='day', y='total_bill', data=tips, size=4, color='black' )
sns.violinplot(x='day', y='total_bill', data=tips)
```

```
#Factorplot method is more general.  You just generically call it and give it a kind
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

# Matrix Plots
Primarily heatmaps
To be a matrix, variables must be present on both columns and rows.

---



```
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```
tips.head()
```

```
#This correlation method puts the data into a matrix format.
tc = tips.corr()
tc
```

```
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```
#Create a pivot table for the flights data.  This puts it into a matrix format.
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```
sns.set_context('poster', font_scale=1)
plt.title = 'Heatmap'
sns.heatmap(fp, cmap='magma', linecolor='black', linewidths=1).set_title('HeatMap')
```

### **Cluster Heat Map**

```
sns.clustermap(fp,cmap='coolwarm', standard_scale=1)
```

# Grids

---



```
iris = sns.load_dataset('iris')
```

```
iris.info()
```

```
import matplotlib.pyplot as plt
```

```
#PairGrid is more customizable than a pairplot
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```
f = sns.FacetGrid(data=tips, col='time' , row='smoker')
plt.figure(figsize=(12,12))
f.map(plt.scatter,'tip','total_bill')
```

# Regression Plots in Seaborn

---



```
sns.set_context('poster')
plt.figure(figsize=(12,12))
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], 
          scatter_kws={'s':100})
```

```
#Add rows and columns of plots with col= and row= instead of facetgrids or hue
sns.lmplot(x='total_bill', y='tip', data=tips,col='sex', row='time', hue='smoker',
          aspect = 0.6, size=8)
```
